# Segmentação baseada em Limiarização(Threshold) 

## Tipos de Limiarização
_, thresh -> o valor do tracinho atras da variavel thresh, é o valor da limiar 

````python
def mostrar(imagem):
    fig = plt.gcf()
    fig.set_size_inches(18,6)
    plt.imshow(imagem, cmap='gray')
    plt.axis('off')
    plt.show()
````

````python
img = cv2.imread('imagens\paisagem01.jpg')
plt.imshow(img)
````

````python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray)
````

````python
val, thresh = cv2.threshold(gray, limiar, 255, cv2.THRESH_BINARY)
````

````python
_, thresh = cv2.threshold(gray, limiar, 255, cv2.THRESH_BINARY)
mostrar(thresh)
````

O comando:

```python
_, thresh = cv2.threshold(gray, limiar, 255, cv2.THRESH_BINARY)
```

é usado para aplicar uma **binarização** à imagem `gray` com um valor de limiar definido por você. Vamos detalhar cada parte:

---

### Detalhamento dos Argumentos:

1. **`gray`**:
   - A imagem de entrada em escala de cinza que você deseja binarizar.

2. **`limiar`**:
   - O valor do limiar (threshold). Qualquer pixel com valor maior ou igual a este será ajustado para 255 (branco), e os demais serão ajustados para 0 (preto).

3. **`255`**:
   - O valor máximo a ser atribuído aos pixels que atendem à condição de limiar (neste caso, 255).

4. **`cv2.THRESH_BINARY`**:
   - Especifica o tipo de binarização:
     - Pixels maiores ou iguais ao valor do limiar, ou seja 180, são ajustados para o valor máximo (255).
     - Pixels menores que o valor do limiar são ajustados para 0.

---

### O Que o Comando Retorna?

1. **`_` (Descarte do valor)**:
   - O primeiro valor retornado pela função é o limiar usado. Ele não é necessário aqui porque você já passou o valor explicitamente (em `limiar`), então é descartado usando o **underscore** `_`.

2. **`thresh`**:
   - A imagem resultante após a binarização, onde os pixels foram ajustados para 0 ou 255 de acordo com o limiar.

---

### O Que Você Verá:
- **Regiões Brancas (255)**: Pixels cuja intensidade foi maior ou igual a `limiar` (127 no exemplo).
- **Regiões Pretas (0)**: Pixels cuja intensidade foi menor que `limiar`.

---

````python
_, thresh = cv2.threshold(gray, limiar, 255, cv2.THRESH_BINARY_INV)
mostrar(thresh)
````

O comando:

```python
_, thresh = cv2.threshold(gray, limiar, 255, cv2.THRESH_BINARY_INV)
```

é uma variação do processo de binarização, mas com a **lógica invertida**. Vamos entender cada parte.

---

### Diferença de **`cv2.THRESH_BINARY_INV`**:
- Este método funciona de forma semelhante ao **`cv2.THRESH_BINARY`**, mas **inverte os valores**:
  - Pixels **maiores ou iguais** ao limiar são ajustados para 0 (preto).
  - Pixels **menores** que o limiar são ajustados para 255 (branco).

---

### Argumentos do Comando:

1. **`gray`**:
   - A imagem de entrada em escala de cinza que será binarizada.

2. **`limiar`**:
   - O valor do limiar para separar os pixels.

3. **`255`**:
   - O valor máximo atribuído aos pixels que atendem à condição inversa do limiar.

4. **`cv2.THRESH_BINARY_INV`**:
   - O tipo de binarização que inverte os resultados do método **`cv2.THRESH_BINARY`**.

---

### Retorno:

1. **`_`**:
   - O limiar usado (não necessário nesse caso, pois já passamos explicitamente).

2. **`thresh`**:
   - A imagem binarizada, onde os valores foram invertidos:
     - Regiões que seriam **brancas** no método **`THRESH_BINARY`** agora são **pretas**.
     - Regiões que seriam **pretas** no método **`THRESH_BINARY`** agora são **brancas**.

---

### Resultado:
- **Regiões Pretas (0)**: Pixels com intensidade maior ou igual ao limiar (180 no exemplo).
- **Regiões Brancas (255)**: Pixels com intensidade menor que o limiar.

---

### Quando Usar `THRESH_BINARY_INV`?
- **Mascaras**: É útil para criar máscaras invertidas.
- **Pré-processamento**: Quando você precisa que o fundo seja branco e o objeto seja preto (ou vice-versa) em tarefas como segmentação ou detecção de bordas.


````python
_, thresh = cv2.threshold(gray, limiar, 255, cv2.THRESH_TRUNC)
mostrar(thresh)
````

O comando:

```python
_, thresh = cv2.threshold(gray, limiar, 255, cv2.THRESH_TRUNC)
```

aplica a técnica de limiarização **truncada** à imagem em escala de cinza (`gray`). Este método funciona de forma diferente dos métodos binários. Vamos entender em detalhes.

---

### Como funciona o **`cv2.THRESH_TRUNC`**:
- **Pixels maiores ou iguais** ao limiar são ajustados para o valor do limiar.
- **Pixels menores** que o limiar permanecem inalterados.

---

### Argumentos:

1. **`gray`**:
   - A imagem de entrada em escala de cinza.

2. **`limiar`**:
   - O valor do limiar que define o ponto de corte para truncar os valores dos pixels.

3. **`255`**:
   - O valor máximo não é utilizado nesse método, mas é necessário passar algo (pode ser 255).

4. **`cv2.THRESH_TRUNC`**:
   - Especifica que a limiarização será truncada.

---

### Retorno:

1. **`_`**:
   - O valor do limiar utilizado (não usado diretamente aqui, já que foi passado explicitamente).

2. **`thresh`**:
   - A imagem resultante após a aplicação do limiar truncado:
     - Para pixels com valor **maior ou igual ao limiar**, o valor é truncado para o limiar.
     - Para pixels com valor **menor que o limiar**, o valor permanece inalterado.

---


### Resultado:
- **Pixels com valores >= `limiar`**:
  - Ajustados para exatamente o valor do limiar (`180` no exemplo).
- **Pixels com valores < `limiar`**:
  - Permanecem com os valores originais.

---

### Quando Usar `THRESH_TRUNC`?

- **Preservar informações**: Quando você precisa reduzir valores de pixel altos (acima do limiar), mas deseja preservar os valores abaixo do limiar.
- **Pré-processamento**: Pode ser útil em detecção de bordas ou segmentação quando você quer destacar áreas de intensidade menor.


````python
_, thresh = cv2.threshold(gray, limiar, 255, cv2.THRESH_TOZERO)
mostrar(thresh)
````

O comando:

```python
_, thresh = cv2.threshold(gray, limiar, 255, cv2.THRESH_TOZERO)
```

aplica o método de limiarização **"TOZERO"**. Este método funciona de maneira diferente das binarizações comuns, pois mantém os valores dos pixels abaixo do limiar e ajusta os valores acima do limiar.

---

### Como funciona o **`cv2.THRESH_TOZERO`**:
- **Pixels com valores maiores ou iguais ao limiar**: Permanecem inalterados.
- **Pixels com valores menores que o limiar**: São ajustados para 0.

---

### Argumentos:

1. **`gray`**:
   - A imagem de entrada em escala de cinza.

2. **`limiar`**:
   - O valor do limiar usado como ponto de corte para zerar os valores dos pixels.

3. **`255`**:
   - Valor máximo, mas não é diretamente utilizado nesse método (é necessário para a assinatura da função).

4. **`cv2.THRESH_TOZERO`**:
   - Especifica que os valores abaixo do limiar serão zerados e os valores acima do limiar permanecerão inalterados.

---

### Retorno:

1. **`_`**:
   - O valor do limiar utilizado (não usado diretamente neste exemplo, pois você já passou o valor explicitamente).

2. **`thresh`**:
   - A imagem resultante após a aplicação do limiar TOZERO:
     - Valores de pixel **maiores ou iguais ao limiar** permanecem inalterados.
     - Valores de pixel **menores que o limiar** são ajustados para **0**.

---

### Resultado:
- **Pixels com valores >= `limiar`**:
  - Permanecem com os valores originais.
- **Pixels com valores < `limiar`**:
  - São ajustados para **0** (preto).

---

### Quando Usar `THRESH_TOZERO`?

- **Segmentação seletiva**:
  - Útil quando você quer destacar regiões de interesse acima de um determinado valor e ignorar outras.
- **Pré-processamento**:
  - Pode ser usado em tarefas de segmentação ou preparação de máscaras para outras operações.

````python
_, thresh = cv2.threshold(gray, limiar, 255, cv2.THRESH_TOZERO_INV)
mostrar(thresh)
````

_, thresh = cv2.threshold(gray, limiar, 255, cv2.THRESH_TOZERO_INV)

aplica a técnica de limiarização **"TOZERO_INV"**. Este método é uma variação do **`cv2.THRESH_TOZERO`**, mas com a lógica invertida.

---

### Como funciona o **`cv2.THRESH_TOZERO_INV`**:
- **Pixels com valores menores que o limiar**: Permanecem inalterados.
- **Pixels com valores maiores ou iguais ao limiar**: São ajustados para **0**.

---

### Argumentos:

1. **`gray`**:
   - A imagem de entrada em escala de cinza.

2. **`limiar`**:
   - O valor do limiar usado como ponto de corte.

3. **`255`**:
   - O valor máximo (não é diretamente utilizado neste método, mas deve ser especificado).

4. **`cv2.THRESH_TOZERO_INV`**:
   - Especifica que os valores acima ou iguais ao limiar serão ajustados para 0, enquanto os valores abaixo permanecem inalterados.

---

### Retorno:

1. **`_`**:
   - O valor do limiar utilizado (não usado diretamente aqui).

2. **`thresh`**:
   - A imagem resultante após a aplicação do limiar TOZERO_INV:
     - Valores de pixel **menores que o limiar** permanecem inalterados.
     - Valores de pixel **maiores ou iguais ao limiar** são ajustados para **0**.

---

### Resultado:
- **Pixels com valores < `limiar`**:
  - Permanecem com os valores originais.
- **Pixels com valores >= `limiar`**:
  - São ajustados para **0** (preto).

---

### Quando Usar `THRESH_TOZERO_INV`?

- **Segmentação seletiva**:
  - Útil quando você quer destacar regiões de interesse abaixo de um valor específico e ignorar as outras.
- **Máscaras para processamentos específicos**:
  - Quando é necessário preservar apenas os valores abaixo de um limiar.
